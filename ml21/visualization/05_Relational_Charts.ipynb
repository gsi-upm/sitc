# Relational charts
Charts for showing correlations.

We have already seen correlation charts in the distribution charts (**lmplot**).

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head(10)
```

### Heatmaps

```python
df = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(df)
```

```python
sns.heatmap(df, annot=True, fmt="d")
```

## Parallel coordinates

```python
from pandas.plotting import parallel_coordinates

df = sns.load_dataset('tips')
df.head(5)
```

```python
fig = plt.figure(figsize=(12, 10))
title = fig.suptitle("Parallel Coordinates", fontsize=18)
fig.subplots_adjust(top=0.93, wspace=0)

pc = parallel_coordinates(df, 
                          'smoker', # class name 
                          ['total_bill', 'tip', 'size'],# numeric columns to show
                          color=('skyblue', 'firebrick'))
```
